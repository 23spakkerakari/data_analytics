```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\pradh\mydata\pandas\datasets\adult.csv')
df
```

```python
#FETCH RANDOM SAMPLES FROM THE DATASET (50%)

df.sample(frac = 0.5, random_state = 100) # will get a different random sample each time
                                          # random_state = any n will keep the same set of random samples each time
                                          
```

```python
#FIND NUM OF NULL VALS

df.isnull().sum()

#Let's use matplotlib to plot this
plt.imshow(df.isnull(), cmap='viridis', aspect = 'auto')
plt.show()
```

```python
df.tail(20)
```

```python
df.isin(['?']).sum()
```

```python
#Let's replace the question marks that we found above

df['workclass'] = df['workclass'].replace('?', np.nan)
df['occupation'] =df['occupation'].replace('?', np.nan)
df['native-country'] = df['native-country'].replace('?', np.nan)
df.tail(20)
```

```python
df.isnull().sum()
```

```python
plt.imshow(df.isnull(), cmap='viridis', aspect = 'auto')
plt.show()
```

```python
#Drop all the missing values

df.isnull().sum() * 100/len(df)
```

```python
df.dropna(how = 'any', inplace = True) # how = 'any' will drop rows with any missing values
df.shape
```

```python
#check and drop for duplicate data

dup = df.duplicated().any()  #without the any(), it returns an nparray, series. the any() will return a t/f val (whether there are dups or nah)
print("Are there any duplicates", dup)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
#DROP THE COLUMNS, EDUCATION-NUM, CAPITAL-GAIN, AND CAPITAL-LOSS

df.drop(['educational-num', 'capital-gain', 'capital-loss' ], axis = 1, inplace = True)
df.columns
```

Univariate Analysis

```python
#FROM HERE, WE WILL BE CONDUCTED UNIVARAITE ANALYSIS ON OUR DATA, to describe data and find patternw ith respect to one variable
```

```python
#What is the distribution of age column

df['age'].describe() 
df['age'].plot(kind = 'hist') #let's plot the distribution of this
```

```python
#Find the total number of people between 17 to 48

df[(df['age'] >= 17) & (df['age'] <= 48)].count()
df[df['age'].between(17, 48)]
```

```python
#Distribution of the workclass column

df['workclass'].describe()
df['workclass'].hist()
```

```python
#how many people have bach/masters degree?/

df[(df['education'] == 'Bachelors') | (df['education'] == 'Masters')].count()
df[df['education'].isin(['Bachelors', 'Masters'])].count()
```

Bivariate Analysis

```python
df.plot(kind = 'box', x = df['age'], y = df['income'])
```

```python
#Let's replace salary values ['<=50k, '>50k'] with 0 and 1

def salary_data(x):
    if(x == '<=50k'):
        return 0
    else:
        return 1

df['encoded_income'] = df['income'].apply(salary_data)
df['encoded_income'].value_counts()
```
